```python
# 创建数据库
import pymysql

db = pymysql.connect("localhost","root","root","bilidata" )

cursor = db.cursor()

# 个人信息库
sql = """CREATE TABLE userinfo (
        id int not null,
        name varchar(20),
        sex int(1),
        birthday varchar(20),
        article int,
        fans int,
        followers int,
        level int
        )"""

# 番剧关注库 uid用户 aid 番剧编号
sql2 = """create table animefollow(
        uid int,
        aid int
        )"""
# 用户关注库 uid用户， ufid关注者编号
sql3 = """create table userfollow(
        uid int,
        ufid int
        )"""

# 番剧信息
# sql4 = """create table animeinfo(
#         aid int,
#         anmae varchar(50)
#         )"""
cursor.execute(sql)
cursor.execute(sql2)
cursor.execute(sql3)
db.close()
```

```python
chrome_options = Options()
iplen = len(iplist)

# 遍历的第n个uuid， 代理ip使用 对iplen取模的代理，
# num = 1

failList = []

for num in range(1,10):
    
    # 设置代理ip
    chrome_options.add_argument("--proxy-server=http://" + iplist[num%iplen][0])
    # 不开启浏览器
    chrome_options.add_argument('--headless')
    # 禁用gpu
    chrome_options.add_argument('--disable-gpu')
    # 模仿浏览器user-agent
    chrome_options.add_argument('user-agent="Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36"')

    driver = webdriver.Chrome(chrome_options=chrome_options)

    url="https://space.bilibili.com/"

    # 读取页面开始try，若有问题直接跳过。
    try:
        driver.get(url+str(num))

        html = driver.page_source

        soup = BeautifulSoup(html, "lxml")

        name = soup.find('span',id='h-name').text
        follows = ''.join(soup.find('a',class_='n-gz').attrs['title'].split(','))
        fans = ''.join(soup.find('a',class_='n-fs').attrs['title'].split(','))
        views = ''.join(soup.find('div',class_='n-bf').attrs['title'].split(','))
        engine2 = create_engine('mysql+pymysql://root:root@localhost:3306/bilidata')
        engine2.execute("insert into userinfo (uuid, name, follows, fans, views) values ('"+str(num)+"','"+name+"','"+follows+"','"+fans+"','"+views+"')")
        print("成功将第",num,"用户信息读入数据库")
    except Exception as e:
        print("第",num,"用户失败")
        failList.append(num)
        print(e)

    driver.close() # 回收资源
```
